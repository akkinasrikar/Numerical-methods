# Exhaustive search

```python
import numpy as np
def function(f,x):
    try:
        return eval(f)
    except:
        return float('inf')
def Exhaustive_search(func,a,b,dx,r):
    deltax=dx
    x1=a
    boolean=0
    it=1
    print(f"Delta x={deltax} ")
    while(boolean==0):
        print(f"iteration {it}")
        x2=x1+deltax
        x3=x2+deltax
        print(f"x1={x1},x2={x2},x3={x3}")
        fx1,fx2,fx3=function(func,x1),function(func,x2),function(func,x3)
        print(f"fx1={round(fx1,r)},fx2={round(fx2,r)},fx3={round(fx3,r)}")
        if(fx1>fx2<fx3):
            print("final")
            print(x1,x3)
            boolean=1
        else:
            it+=1
            x1=x1+deltax
```

# Bounding phase method

```python
def func(f,x):
    try:
        return eval(f)
    except:
        return float('inf')
def bpm(f,a,d):
    x0,k,d=a,0,d
    x=[]
    print(f"func(f,{x0-abs(d)})  func(f,{x0})  func(f,{x0+abs(d)})")
    print(f"{func(f,x0-abs(d))}  {func(f,x0)}  {func(f,x0+abs(d))}")
    if func(f,x0-abs(d))>=func(f,x0)>=func(f,x0+abs(d)):
        print("func(f,x0-abs(d))>=func(f,x0)>=func(f,x0+abs(d))")
        print("d is positive")
        d=+(d)
    elif func(f,x0-abs(d))<=func(f,x0)<=func(f,x0+abs(d)):
        print("func(f,x0-abs(d))<=func(f,x0)<=func(f,x0+abs(d))")
        print("d is negative")
        d=-(d)
    else:
        print("choose different values")
    xp=x0+(2**k)*d
    x=[x0,xp]
    itr=0
    print(f"x{itr}={x[0]}")
    itr+=1
    print(f"x{itr}={x[1]}")
    while(func(f,xp)<func(f,x0)):
        x0=xp
        k=k+1
        xp=x0+(2**k)*d
        itr+=1
        print(f"x{itr}={xp}")
        print(f"f({x0})={func(f,x0)} and f({xp})={func(f,xp)}")
        x.append(xp)
        #print(x0,xp)
    print(x)
    print(f"final interval is {x[-3]} and {x[-1]}")
```

# interval halving method

```python
def func(f,x):
    return eval(f)
def mod(n):
    if n<0:
        return -n
    return n
def interval_halving_method(f,a,b,e):
    xm=(a+b)/2
    l=b-a
    print(f"l={l}")
    itr=1
    while(mod(l)>e):
        print(f"iteration {itr}")
        x1,x2=a+l/4,b-l/4
        fxm=func(f,xm)
        print(f"x1={x1}  x2={x2}")
        fx1,fx2=func(f,x1),func(f,x2)
        print(f"fx1={fx1}  fx2={fx2}  fxm={fxm}")
        if fx1<fxm:
            b=xm;
            xm=x1;
            print(f"b={xm} xm={x1}")
        elif fx2<fxm:
            a=xm;
            xm=x2;
            print(f"a={xm} xm={x2}")
        else:
            a=x1;
            b=x2;
            print(f"a={a} b={b}")
        itr+=1
        l=b-a
        print(f"l=b-a={l}")
    print(a,b)
```

```python
interval_halving_method('x**2+(54/x)',0,5,0.5)
```

# Fibonacci Search Method

```python
def fibo(n):
    a,b=0,1
    ch=2
    while(ch<=n+1):
        c=a+b
        a,b=b,c
        ch+=1
    return b
def func(f,x):
    return eval(f)
def fibo_search(f,a,b,no,r,k=None):
    if k is None:
        k=2
    a,b=a,b
    l=b-a
    print(f"l={l}")
    while(k<=no):
        print(f"k={k}")
        print(f"L{k}*=(F{no-k+1}/F{no+1})*l")
        print(f"L{k}*=({fibo(no-k+1)}/{fibo(no+1)})*{l}")
        lk=(fibo(no-k+1)/fibo(no+1))*l
        x1,x2=a+lk,b-lk
        print(f"L{k}*={round(lk,r)} x1={round(x1,r)}  x2={round(x2,r)}")
        fx1,fx2=func(f,x1),func(f,x2)
        if(fx1>fx2):
            a=x1
            b=b
            print(f"a={round(x1,r)}  b={round(b,r)}")
        if(fx1<fx2):
            b=x2
            a=a
            print(f"a={round(a,r)}  b={round(x2,r)}")
        k=k+1
        #print(a,b)
        #print(fx1,fx2)
```

# Golden search Method 

```python
def func(f,x):
    return eval(f)
def wfunc(f,x,a,b):
    return eval(f)
def mod(n):
    if n<0:
        return -n
    return n
def golden_search(f,w,a,b,e,r):
    lw=1
    k=1
    fu='(x-a)/(b-a)'
    aw=wfunc(fu,a,a,b)
    bw=wfunc(fu,b,a,b)
    itr=1
    print(aw,bw)
    while(mod(lw)>e):
        print(f"iteration {itr}")
        w1=aw+(0.618*lw)
        w2=bw-(0.618*lw)
        fw1=func(w,w1)
        fw2=func(w,w2)
        print(f"w1 {round(w1,r)} and w2 {round(w2,r)}")
        print(f"fw1 {round(fw1,r)} fw2 {round(fw2,r)}")
        if fw1<fw2:
            aw=w2
            lw=bw-aw
            #print(lw)
        elif fw2<fw1:
            bw=w1
            lw=bw-aw
            #print(lw)
        itr+=1
        print(f"aw={round(aw,r)}  bw={round(bw,r)}  lw={round(lw,r)}")
```

# 1st way

```python
import numpy as np
def func(f,x):
    return eval(f)
def nrp(x,fd,fdd,e):
    k=1
    while True:
        print(f"iteration {k}")
        print(f"x{k}--> {round(x,4)}")
        print(f"f'(x{k})={round(func(fd,x),4)} and f''(x{k})={round(func(fdd,x),4)}")
        x=x-func(fd,x)/func(fdd,x)
        print(f"x{k+1}--> {round(x,4)}")
        print(f"f'(x{k+1})={round(func(fd,x),4)}")
        fdx=func(fd,x)
        if abs(fdx)<e:
            break
        k=k+1
```

```python
import sympy as sp
x=sp.Symbol('x')
xdash=sp.diff(x**2+54/x)
xdoudash=sp.diff(xdash)
print("1st derivative",xdash,"2st derivative",xdoudash)
```

# 2nd way

```python
def func(f,x):
    return eval(f)
def nrp(x,f,fd,e):
    k=1
    while True:
        if abs(x)>0.01:
            dx=0.01*abs(x)
        else:
            dx=0.0001
        x=x-((func(f,x+dx)-func(f,x-dx))/2*dx)/((func(f,x+dx)-2*func(f,x)+func(f,x-dx))/(dx**2))
        fdx=func(fd,x)
        print(f"iteration {k}")
        print(f"x--> {x}")
        print(f"fdx--> {fdx}")
        if abs(fdx)<e:
            break
        k=k+1
```

# Bisection method

```python
import numpy as np
def func(fc,x):
    return(eval(fc))
def bisection_method(fc,a,b,r,e):
    z=round((a+b)/2,r)
    print(f"f'({a})={func(fc,a)} and f'({b})={func(fc,b)}")
    it=1
    while True:
        print(f"iteration {it}")
        print(f"z={z}")
        fz=func(fc,z)
        print(f"f'({z})={round(func(fc,z),r)}")
        if fz<0:
            a=z
            print(f"f'({z})<0 so x1={a}")
        elif fz>0:
            b=z
            print(f"f'({z})>0 so x2={b}")
        print(f"x1={a} and x2={b}")
        if abs(fz)<e:
            break
        z=round((a+b)/2,r)
        it=it+1
```

```python
bisection_method("2*x - 54/x**2",1,5,3,1e-3)
```

# Secant method

```python
import numpy as np
def func(fc,x):
    return(eval(fc))
def secant_method(fc,a,b,r,e):
    z=b-(func(fc,b)/((func(fc,b)-func(fc,a))/(b-a)))
    print(f"f'({a})={func(fc,a)} and f'({b})={func(fc,b)}")
    it=1
    while True:
        print(f"iteration {it}")
        print(f"f'({round(a,r)})={round(func(fc,a),r)} and f'({round(b,r)})={round(func(fc,b),r)}")
        print(f"z={round(z,r)}")
        fz=func(fc,z)
        print(f"f'({round(z,r)})={round(func(fc,z),r)}")
        if fz<0:
            a=z
            print(f"f'({round(z,r)})<0 so x1={round(a,r)}")
        elif fz>0:
            b=z
            print(f"f'({round(z,r)})>0 so x2={round(b,r)}")
        print(f"x1={round(a,r)} and x2={round(b,r)}")
        if abs(fz)<e:
            break
        z=b-(func(fc,b)/((func(fc,b)-func(fc,a))/(b-a)))
        it=it+1
```

# Boxs Evolutionary Optimization Method

```python
import numpy as np
def func(f,x,y):
    return eval(f)
def md(l):
    return abs(np.sqrt(l[0]**2+l[1]**2))
def BEOM(f,x0,d,e):
    it=1
    b=1
    while b==1:
        d=d
        xbar=x0
        t=d[0]/2
        if md(d)>e:
            x1=[x0[0]-t,x0[1]-t]
            x2=[x0[0]+t,x0[1]-t]
            x3=[x0[0]-t,x0[1]+t]
            x4=[x0[0]+t,x0[1]+t]
        else:
            b=0
        fx0,fx1=func(f,x0[0],x0[1]),func(f,x1[0],x1[1])
        fx2,fx3=func(f,x2[0],x2[1]),func(f,x3[0],x3[1])
        fx4=func(f,x4[0],x4[1])
        lfxv=[x0,x1,x2,x3,x4]
        lfx=[fx0,fx1,fx2,fx3,fx4]
        m=min(lfx)
        l=lfx.index(m)
        xbar=lfxv[l]
        print(f"iteration {it}")
        print(f"delta={d}")
        print(f"t={t}")
        print(f"||delta||={md(d)}")
        print("points from x0 to x4")
        print(lfxv)
        print("f values from x0 to x4")
        print(lfx)
        print(f"xbar={xbar}")
        if xbar!=x0:
            x0=xbar
        else:
            d=[d[0]-t,d[1]-t]
            print("delta is changing")
        it=it+1
```
